## Importing necessary libraries

```python
import tensorflow as tf
from tensorflow.keras.datasets import mnist
```

```python
# Load the dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

## Plotting the Images

```python
import matplotlib.pyplot as plt
import numpy as np

random_idx = np.random.randint(0, len(x_test)) # for getting any random digit images

random_image = x_test[random_idx]              # getting the random image index and label
random_label = y_test[random_idx]

plt.imshow(random_image, cmap='gray')          # ploting the images using imshow
plt.title(f"Label: {random_label}")
plt.show()
```

## Normalization (converting to range (0 to 1))

Normalization in deep learning (DL) is the process of transforming data into a standard scale, usually between 0 and 1.

```python
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255
```

## Flatten Image (Reshape the images)

Flattening is a technique that is used to convert multi-dimensional arrays into a 1-D array, it is generally used while feeding the 1-D array to the classification model.

```python
x_train = x_train.reshape((x_train.shape[0], 28 * 28))
x_test = x_test.reshape((x_test.shape[0], 28 * 28))
```

## One - Hot Encoding

Converts the categorical column to numerical into different columns.

```python
from tensorflow.keras.utils import to_categorical
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)
```

## Activation Functions

#### An Activation Function decides whether a neuron should be activated or not.
#### This means that it will decide whether the neuron's input to the network is important or not in the process of prediction using simpler mathematical operations.

```python
activation_functions = ['sigmoid', 'tanh', 'relu', 'swish']
```

image.png

Think of tanh as a function that squashes any input into a small range between -1 and 1, helping to stabilize the outputs of a neural network layer.

image.png

Imagine ReLU as a function that "cuts off" any negative values and only lets positive values pass through unchanged. It's like a "gate" that blocks negatives and lets positives flow freely.

image.png

Think of the sigmoid function as an "S-shaped" curve that smoothly transitions from 0 to 1. Small negative inputs produce outputs close to 0, small positive inputs produce outputs close to 1, and inputs around 0 produce outputs near 0.5.

image.png

Swish: Developed by Google, it's a blend of ReLU and sigmoid

## Designing a Simple Neural Network

image.png

```python
from tensorflow.keras import layers, models
from tensorflow.keras.activations import swish
```

```python
accuracy_dict = {}    # to store the values
model_dict = {}

for activation in activation_functions:
    print(f'\nTesting with activation function: {activation}')

    # Sequential model
    model = models.Sequential()

    if activation == 'swish':                                                        # Swish activation
        model.add(layers.Dense(128, activation=swish, input_shape=(28 * 28,)))
    else:
        model.add(layers.Dense(128, activation=activation, input_shape=(28 * 28,)))  # Other activations

    model.add(layers.Dense(10, activation='softmax'))  # Output layer with softmax activation



    model.compile(optimizer='sgd',                 # Stochastic Gradient Descent
                  loss='categorical_crossentropy', # Used as loss function for multi-class classification model
                  metrics=['accuracy'])            # Accuracy metric




    print('Training model...')
    history = model.fit(x_train, y_train,
              epochs=5,                    # During one epoch, the model processes all 60,000 training samples
              batch_size=64,               # steps_per_epoch = 60,000/64 = 937.5
              verbose=1)                   # verbose (0,1,2)



    print('\nEvaluating model...')
    test_loss, test_acc = model.evaluate(x_test, y_test, verbose=1)
    print(f"Test Accuracy: {test_acc:.4f}")
    print(f"Test Loss: {test_loss:.4f}")


    accuracy_dict[activation] = test_acc
    model_dict[activation] = model
```

## Accuracies of different activation functions

```python
for activation, accuracy in accuracy_dict.items():
    print(f'Accuracy with {activation}: {accuracy * 100:.2f}%')
```

```python
plt.figure(figsize=(10, 6))
plt.bar(accuracy_dict.keys(), accuracy_dict.values(), color=['lightblue', 'lightgreen', 'lightcoral', 'plum'])
plt.title('Comparison of Activation Function Accuracies')
plt.xlabel('Activation Function')
plt.ylabel('Accuracy')
plt.ylim(0.8, 1)

plt.show()
```

##### Here, above bar graph displays that tanh & relu functions have the highest accuracies 91% & 92%, followed by sigmoid & swish with 91% & 88%.

## Prediction of random image with different activation functions

```python
def predict_random_image(model):
    random_idx = np.random.randint(0, len(x_test))
    random_image = x_test[random_idx]
    random_label = np.argmax(y_test[random_idx])

    predicted_label = np.argmax(model.predict(random_image.reshape(1, 28 * 28)))

    plt.figure(figsize=(2, 4))
    plt.imshow(random_image.reshape(28, 28), cmap='gray')
    plt.title(f"Actual: {random_label}, Predicted: {predicted_label}")
    plt.axis('off')
    plt.show()
```

```python
for activation in activation_functions:
    print(f'\nPredicting with activation function: {activation}')
    predict_random_image(model_dict[activation])
```

##### Here, we can see that all the activation functions have predicted the right digit classification due to high accuracies.

#### Based on the results, the activation function ReLU performed the best on the MNIST classification task with an accuracy of 92.05%, closely followed by Tanh with 91.84%, and Swish with 91.53%. The Sigmoid activation function showed the lowest performance, achieving an accuracy of 88.26%.

#### In conclusion, while all activation functions performed reasonably well, ReLU demonstrated the highest accuracy, making it the most effective activation function for this specific task.
